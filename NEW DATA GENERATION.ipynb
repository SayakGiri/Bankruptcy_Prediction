## Generating New Data to check effectiveness of the model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from scipy import stats 
from scipy.stats import iqr
import seaborn as sns
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("data.csv")
df = df.rename(columns={"Bankrupt?": "Bankrupt"})
df = df.drop(df.columns[[94]], axis=1)
df.head()
```

```python
p=df[df['Bankrupt']==1]
```

```python
q=df[df['Bankrupt']==0]
```

**Generating some new 1's**

```python
a1=q.iloc[50:600,]
pa1=pd.concat([p,a1])
```

```python
y1=pa1['Bankrupt']
x1=pa1.iloc[:,1:95]
sm = SMOTE(random_state=481)
x1, y1 = sm.fit_resample(x1, y1)
```

```python
pa2=pd.DataFrame(y1,columns=['Bankrupt'])
for i in range(0,94):
    pa2[x1.columns[i]]=x1.iloc[:,i]
```

**Generating some new 0's**

```python
q1=q.iloc[1:50,]
```

```python
pq1=pd.concat([p,q1])
```

```python
y2=pq1['Bankrupt']
x2=pq1.iloc[:,1:95]
```

```python
sm = SMOTE(random_state=481)
x2, y2 = sm.fit_resample(x2, y2)
```

```python
pq2=pd.DataFrame(y2,columns=['Bankrupt'])
for i in range(0,94):
    pq2[x2.columns[i]]=x2.iloc[:,i]
```

```python
paq=pd.concat([pa2,pq2])
paq
```
